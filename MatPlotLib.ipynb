```python
from matplotlib import pyplot as plt
```

```python
X=[1,2,3,4,5,6,7,8]
```

```python
Y=[2,3,5,6,6,8,9,10]
```

```python
plt.plot(X,Y,'k')
plt.show()
```

```python
plt.plot(X,Y,'r')
plt.show()
```

```python
plt.plot(X,Y,'bo')
plt.show()
```

```python
plt.plot(X,Y,'*')
plt.show()
```

```python
Z=[3,5,6,0,7,9,11,12]
plt.plot(X,Z,'ro')
plt.show()
```

```python
plt.plot(X,Y,'b>')
plt.plot(X,Z,'r*')
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend(["Y","Z"])
plt.title("XvsY")
plt.show()
```
