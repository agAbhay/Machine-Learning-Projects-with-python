```python
sb.lmplot(x="Defense",y="Attack",data=bunny, \
          scatter_kws={'marker':'*','color':'red'}, \
         line_kws={"linewidth":3,"color":"blue"})
plt.show()
```

```python
sb.lmplot(x="Defense",y="Attack",data=bunny, \
          scatter_kws={'marker':'*','color':'red'}, \
         line_kws={"linewidth":3,"color":"blue"},hue="Stage")
plt.show()
```

```python
sb.lmplot(x="Defense",y="Attack",data=bunny, \
     hue="Stage",markers=["*","o",">"])
plt.show()
```

```python
sb.countplot(x="Type 1",data=bunny)
plt.show()
```

```python
sb.countplot(x="Type 1",data=bunny,hue="Legendary")
plt.xticks(rotation=45)
plt.show()
```
