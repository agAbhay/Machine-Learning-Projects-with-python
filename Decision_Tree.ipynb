```python
import pandas as pd
```

```python
bunny=pd.read_csv(r"F:\New Datasets\Ml_Data\Titanic_Data.csv",index_col="PassengerId")
```

```python
bunny
```

```python
Titanic_Data=bunny[['Pclass','Sex','Age','SibSp','Parch','Fare','Survived']]
```

```python
Titanic_Data
```

```python
Titanic_Data.head(1)
```

```python
Titanic_Data["Sex"]=Titanic_Data["Sex"].map({"male":0,"female":1})
```

```python
Titanic_Data
```

```python
Titanic_Data.describe()
```

```python
#drop the nan values,true is used to update
Titanic_Data.dropna(inplace=True)
```

```python
Titanic_Data
```

```python
Titanic_Data.describe()
```

```python
x_feature=Titanic_Data.drop("Survived",axis=1)
y_target=Titanic_Data["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_feature,y_target,test_size=.30,random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
trainer=DecisionTreeClassifier(criterion="entropy")
```

```python
learner=trainer.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#x_test.....y_test
ya=y_test
yp=learner.predict(x_test)
acc=accuracy_score(ya,yp)*100
acc
```

```python
from sklearn import tree
```

```python
import pydotplus
from IPython.display import Image
```

```python
dot_data=tree.export_graphviz(trainer,out_file=None,feature_names=x_feature.columns)
```

```python
graph=pydotplus.graph_from_dot_data(dot_data)
```
