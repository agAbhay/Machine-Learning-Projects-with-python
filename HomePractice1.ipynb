```python
from sklearn import datasets
```

```python
bunny=datasets.load_iris()
```

```python
bunny
```

```python
bunny.keys()
```

```python
bunny.target
```

```python
bunny.data
```

```python
ya=bunny.data
yp=bunny.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,ytrain,xtest,ytest=train_test_split(ya,yp,test_size=.30,random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
teacher=KNeighborsClassifier()
```

```python
learner=teacher.fit(xtrain,xtest)
```

```python
yaa=ytest
ypp=learner.predict(ytrain)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(yaa,ypp)*100
```
