```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train=X_Input[:350]
X_test=X_Input[350:]
Y_train=Y_output[:350]
Y_test=Y_output[350:]
```

```python
teachergrass=GaussianNB()
teacherber=BernoulliNB()
teacherMul=MultinomialNB()
teacherknn=KNeighborsClassifier()
```

```python
learnergrass=teachergrass.fit(X_train,Y_train)
learnerknn=teacherknn.fit(X_train,Y_train)
learnerber=teacherber.fit(X_train,Y_train)
learnerMul=teacherMul.fit(X_train,Y_train)
```

```python
Ya=Y_test
Ypber=learnerber.predict(X_test)
YpMul=learnerMul.predict(X_test)
YpGrass=learnergrass.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accber=accuracy_score(Ya,Ypber)*100
accmul=accuracy_score(Ya,YpMul)*100
accGass=accuracy_score(Ya,YpGrass)*100
```

```python
table=kg.DataFrame({"accGass":[accGass],"accmul":[accmul],"accber":[accber]})
```
