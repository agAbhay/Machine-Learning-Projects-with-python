```python
from matplotlib import pyplot as plt
```

```python
X=[1,2,3,4,5,6,7,8]
Y=[2,3,5,6,6,8,9,10]
Z=[3,5,6,0,7,9,11,12]
```

```python
plt.plot(X,Y)
plt.plot(X,Z,'r')
plt.show()
```

```python
plt.plot(X,Y,'r')
plt.plot(X,Z,'r')
plt.show()
```

```python
plt.plot(X,Y,'g')
plt.plot(X,Z,'r')
plt.show()
```

```python
plt.plot(X,Y,'bo')
plt.plot(X,Z,'r')
plt.show()
```

```python
plt.plot(X,Y,'b>')
plt.plot(X,Z,'r*')
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend(["Y","Z"])
plt.title("Population")

plt.show()
```
