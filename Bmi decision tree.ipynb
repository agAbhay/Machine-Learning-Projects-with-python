```python
b_feature=bunny.iloc[:,0:3]
b_target=bunny["Index"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(b_feature,b_target,test_size=.30,random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
trainer=DecisionTreeClassifier(criterion="entropy")
```

```python
learner=trainer.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#x_test.....y_test
ya=y_test
yp=learner.predict(x_test)
acc=accuracy_score(ya,yp)*100
acc
```
