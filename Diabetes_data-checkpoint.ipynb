```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
teacher=KNeighborsClassifier()
```

```python
Ya=Y_test
Yp=learner.predict(X_test)
```

```python
Yalist=list(Ya)
Yplist=list(Yp)
```

```python
table=kg.DataFrame({"Ya":Yalist,"Yp":Yplist})
```
